```python
import pandas as pd
```

```python
df = pd.read_csv('data_raw\archive.csv', 
                 sep=';', encoding='latin-1', header=1)

df['Hora UTC'] = pd.to_datetime(df['Hora UTC'].str[:4], format='%H%M')
df['Hora UTC'] = df['Hora UTC'].dt.time
df['Data'] = pd.to_datetime(df['Data'], format='%d/%m/%Y')
df['data'] = pd.to_datetime(df['Data'].astype(str) + ' ' + df['Hora UTC'].astype(str))

df.rename(columns={'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'preciptacao',
                   'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)': 'pressao_atm',
                   'PRESSÃO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)': 'pressao_atm_max',
                   'PRESSÃO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)': 'pressao_atm_min',
                   'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)': 'temperatura',
                   'TEMPERATURA DO PONTO DE ORVALHO (°C)': 'temperatura_orvalho',
                   'TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)': 'temperatura_max',
                   'TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)': 'temperatura_min',
                   'TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (°C)': 'temperatura_orvalho_max',
                   'TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (°C)': 'temperatura_orvalho_min',
                   'UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)': 'umidade_rel_ar_max',
                   'UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)': 'umidade_rel_ar_min',
                   'UMIDADE RELATIVA DO AR, HORARIA (%)': 'umidade_rel_ar',
                   'VENTO, VELOCIDADE HORARIA (m/s)': 'velocidade_vento'}, inplace=True)

df['preciptacao'] = df['preciptacao'].str.replace(',' , '.').astype(float)
df['pressao_atm'] = df['pressao_atm'].str.replace(',' , '.').astype(float)
df['pressao_atm_max'] = df['pressao_atm_max'].str.replace(',' , '.').astype(float)
df['pressao_atm_min'] = df['pressao_atm_min'].str.replace(',' , '.').astype(float)
df['temperatura'] = df['temperatura'].str.replace(',' , '.').astype(float)
df['temperatura_orvalho'] = df['temperatura_orvalho'].str.replace(',' , '.').astype(float)
df['temperatura_max'] = df['temperatura_max'].str.replace(',' , '.').astype(float)
df['temperatura_min'] = df['temperatura_min'].str.replace(',' , '.').astype(float)
df['temperatura_orvalho_max'] = df['temperatura_orvalho_max'].str.replace(',' , '.').astype(float)
df['temperatura_orvalho_min'] = df['temperatura_orvalho_min'].str.replace(',' , '.').astype(float)    
df['velocidade_vento'] = df['velocidade_vento'].str.replace(',' , '.').astype(float)

df = df[['preciptacao', 'pressao_atm', 'pressao_atm_max', 'pressao_atm_min',
       'temperatura', 'temperatura_orvalho', 'temperatura_max',
       'temperatura_min', 'temperatura_orvalho_max', 'temperatura_orvalho_min',
       'umidade_rel_ar_max', 'umidade_rel_ar_min', 'umidade_rel_ar',
       'velocidade_vento', 'data']]

df
```

```python
df_clima = pd.read_parquet('data\df_clima_sp.parquet')
df_clima = pd.concat([df_clima, df])
```

```python
df_clima.to_parquet('data\df_clima_sp.parquet')
```
